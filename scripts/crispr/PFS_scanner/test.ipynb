```python
def read_rna_sequence_from_file(file_path):
    with open(file_path, 'r') as file:
        rna_sequence = file.read().strip()
    return rna_sequence


def find_target_sequences(rna_sequence):
    target_sequences = []
    #define pattern
    pattern = re.compile(r'[ACGU]AAA[ACGU]')
    
    for i in range(len(rna_sequence) - 28):
        candidate = rna_sequence[i:i+29]
        if pattern.fullmatch(candidate[-5:]):
            target_sequences.append([candidate, i])
    
    return target_sequences
```
